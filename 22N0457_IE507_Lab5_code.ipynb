```python
d.fillna(0)
```

```python
da=np.array(d.fillna(0))
```

```python
data=da[:,1:24]
```

```python
route=np.array(d.keys()[1:24])
```

```python
index={}
z=0
for i in route:
  index[i]=z
  z+=1
```

```python
def check_intersection(a,b):
  if data[index[str(a)]][index[str(b)]]==1:
    return True
  else:
    return False
```

```python
colour=['white','red','blue','green','black','orange','yellow','pink','brown','purple','gray','golden','navy','maroon']
colour1=colour.copy()
```

```python
route3=list(route)
for k in range(len(colour)):
  colour[k]=[]
  colour[k].append(str(route3[0]))
  route3.remove(str(route3[0]))
  for i in route3:
    if sum(check_intersection(str(i),str(j)) for j in colour[k]) == 0:
      colour[k].append(i)
      route3.remove(i)
  if len(route3)==0:
        break
```

```python
for i in range(len(colour)):
  print(colour1[i]+'  '+'colour is used for',colour[i])
```
